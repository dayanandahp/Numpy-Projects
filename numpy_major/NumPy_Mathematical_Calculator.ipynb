```python
import numpy as np

print("Numpy Mathematical Calculator")

# Input handling
try:
    arr1 = np.array(list(map(float, input("Enter the array1 values (comma separated): ").split(','))))
    arr2 = np.array(list(map(float, input("Enter the array2 values (comma separated): ").split(','))))
except ValueError:
    print("Enter valid numeric inputs!")
    exit()

while True:
    print("\nChoose the operation:")
    print("1. Arithmetic Operations")
    print("2. Statistical Operations")
    print("3. Array Manipulation")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        print("\n--- Arithmetic Operations ---")
        try:
            if len(arr1) != len(arr2):
                raise ValueError("Arrays must have the same length for arithmetic operations.")
            print(f"Addition of two arrays: {arr1 + arr2}")
            print(f"Subtraction of two arrays: {arr1 - arr2}")
            print(f"Division of two arrays: {arr1 / arr2}")
            print(f"Floor division of two arrays: {arr1 // arr2}")
        except Exception as e:
            print(f"Error: {e}")

    elif choice == '2':
        print("\n--- Statistical Operations ---")
        try:
            arr = np.concatenate([arr1, arr2])
            print(f"Concatenated array: {arr}")
            print(f"Max element: {np.max(arr)}")
            print(f"Min element: {np.min(arr)}")
            print(f"Sum of elements: {np.sum(arr)}")
            print(f"Product of elements: {np.prod(arr)}")
            print(f"Mean: {np.mean(arr)}")
            print(f"Median: {np.median(arr)}")
            print(f"Rounded values: {np.round(arr)}")
            print(f"Standard deviation: {np.std(arr)}")
        except Exception as e:
            print(f"Error: {e}")

    elif choice == '3':
        print("\n--- Array Manipulation ---")
        try:
            arr = np.concatenate([arr1, arr2])
            try:
                reshape = arr.reshape(2, -1)
                print(f"Reshaped array:\n{reshape}")
            except ValueError:
                print("Cannot reshape array due to incompatible dimensions.")
            print(f"Sorted array: {np.sort(arr)}")
            print(f"Reversed array: {arr[::-1]}")
            print(f"Unique values: {np.unique(arr)}")
            print(f"First and last element: {(arr[0], arr[-1])}")
            mean_val = np.mean(arr)
            print(f"Values greater than mean ({mean_val}): {arr[arr > mean_val]}")
        except Exception as e:
            print(f"Error: {e}")

    elif choice == '4':
        print("\nThank you for using the NumPy Calculator!")
        break

    else:
        print("Invalid choice! Please enter a number between 1 and 4.")
```
